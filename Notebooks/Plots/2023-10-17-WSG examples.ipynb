```python
import pandas as pd
import numpy as np
import pickle
from ssoc_autocoder.combined_model import SSOCAutoCoder
```

```python
with open("Models/autocoder.pkl", "rb") as inp:
    model = pickle.load(inp)
```

```python
data = pd.read_csv("/Users/dsaid/Downloads/output_wsg_jobseeker.csv")
```

```python
data.head()
```

```python
result_list = []
confidence_list = []
```

```python
for index, row in data.iterrows():
    model_results = model.predict_lite(title=row['job_title'])
    prediction = model_results['prediction'][0]
    confidence = model_results['confidence'][0]
    result_list.append(prediction)
    confidence_list.append(confidence)
```

```python
data['title_only_pred'] = result_list
data['title_only_conf'] = confidence_list
```

```python
data.to_csv("/Users/dsaid/Downloads/output_wsg_jobseeker_w_pred.csv", index=False)
```

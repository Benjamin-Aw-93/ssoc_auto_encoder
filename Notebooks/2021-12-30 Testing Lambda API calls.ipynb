```python
import requests
import pandas as pd
```

```python
# Reading in data

df = pd.read_csv("../Data/Processed/MCF_Training_Set_Full.csv")
```

```python
output = []

for text in df['description'][0:100]:
    
    parameters = {'text': text}
    r = requests.get("https://zjkg0alc26.execute-api.ap-southeast-1.amazonaws.com/default/test_cleaning_data", params=parameters)\
    
    if r.status_code == 200:
        output.append(r.json())
    else:
        while r.status_code != 200:
            parameters = {'text': text}
            r = requests.get("https://zjkg0alc26.execute-api.ap-southeast-1.amazonaws.com/default/test_cleaning_data", params=parameters)\
    
        
```
